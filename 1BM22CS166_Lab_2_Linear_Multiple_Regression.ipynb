```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/canada_per_capita_income.csv")
```

```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train,Y_train)
```

```python
y_pred = reg.predict(X_test)
print(y_pred)
print(Y_test)
```

```python

import matplotlib.pyplot as plt
plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, reg.predict(X_train), color = 'blue')
plt.title('Income vs Year (Training set)')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()

plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_train, reg.predict(X_train), color = 'blue')
plt.title('Income vs Year (Test set)')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()
```

```python
df1 = pd.read_csv("/content/salary.csv")
df1.dropna(inplace=True)
```

```python
X1 = df1.iloc[:,:-1].values
Y1 = df1.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1,Y1,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression()
reg1.fit(X1_train,Y1_train)
```

```python

plt.scatter(X1_train, Y1_train, color = 'red')
plt.plot(X1_train, reg1.predict(X1_train), color = 'blue')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

plt.scatter(X1_test, Y1_test, color = 'red')
plt.plot(X1_train, reg1.predict(X1_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
